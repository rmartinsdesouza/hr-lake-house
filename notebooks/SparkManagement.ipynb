```python
import sys
import os
from pyspark.sql import SparkSession
from delta import *
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages io.delta:delta-core_2.12:2.4.0 --conf spark.sql.extensions=io.delta.sql.DeltaSparkSessionExtension --conf spark.sql.catalog.spark_catalog=org.apache.spark.sql.delta.catalog.DeltaCatalog pyspark-shell'
sys.path.append('/content/drive/MyDrive/Portifolio/hr-lake-house/dataset')
```

```python
#Create Session spark
spark = SparkSession.builder.appName("jobSparkManagement") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")\
    .getOrCreate()
```

```python
df = spark.read.format("csv")\
  .option("header", "true")\
  .option("inferSchema", "true")\
  .load("/content/drive/MyDrive/Portifolio/hr-lake-house/dataset/countries.csv")
```

```python
df.show()
```

```python
df.write.format("parquet").option("mode", "overwrite").save("/content/drive/MyDrive/Portifolio/hr-lake-house/notebooks/tb.delta")
```
